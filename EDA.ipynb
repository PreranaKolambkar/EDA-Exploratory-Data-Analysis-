```python
import scipy.stats as stats
```

```python
#function to detect outlieers
```

```python
def detect_outliers_z_score(df, threshold=2):  #standard tgreshold value for z_score
   z_scores = stats.zscore(df)  # Calculate Z-scores for all columns in the dataframe
   outliers = abs(z_scores) > threshold # Detect outliers: if the absolute value of Z-score is greater than the threshold
   columns_with_outliers = df.columns[(outliers).any(axis=0)].tolist() 
    # identifies the columns where any row has an outlier 
   return columns_with_outliers


outlier_columns = detect_outliers_z_score(df)
print( outlier_columns)
```
